```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('traffic_accident.csv')
```

```python
df.head()
```

```python
print(df.columns)
```

```python
print(df.info())
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Age_band_of_driver', data=df, palette='viridis')
plt.title('Distribution of Age of the driver in Accidents')
plt.xlabel('Age')
plt.ylabel('Number of Accidents')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Educational_level', data=df, palette='magma')
plt.title('Distribution of Education in Accidents')
plt.xlabel('Education Level')
plt.ylabel('Number of Accidents')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Time', data=df, palette='plasma')
plt.title('Distribution of Time in Accidents')
plt.xlabel('Time')
plt.ylabel('Number of Accidents')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Weather_conditions', data=df, palette='magma')
plt.title('Distribution of Weather Conditions in Accidents')
plt.xlabel('Weather Conditions')
plt.ylabel('Number of Accidents')
plt.show()
```
